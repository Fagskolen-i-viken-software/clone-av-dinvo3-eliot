```python
# quick sort algoritme basert på redursjon 

indent = " "
def quick_sort(arr, indent= "..."):
    
    
    if len(arr) <= 1:
        return arr  # En liste med 0 eller 1 element er allerede sortert
    
    pivot  = arr[len(arr) // 2]  # Velger midterste element som pivot
    left = []
    right = []
    equal = []

    print("Pivot:", pivot)

    for num in arr: 
        if num < pivot:
            print(indent, "legger til", num, "i venstre")
            left.append(num)
        elif num > pivot:
            print(indent, "legger til", num, "i høyre")
            right.append(num)
        else:
            print(indent, "legger til", num, "i lik")
            equal.append(num)

    indent += "..."
    # Rekursiv sortering av venstre og høyre del
    left = quick_sort(left, indent=indent)
    right = quick_sort(right, indent=indent)

    # Kombinerer de sorterte delene
    return left + equal + right


list_to_sort = [3, 6, 8, 10, 1, 2, 1, 100, 34, 23, 54, 2]
print(len(list_to_sort), "elementer å sortere")
sorted_list = quick_sort(list_to_sort)
print("Sorted list:", sorted_list)
    
```
